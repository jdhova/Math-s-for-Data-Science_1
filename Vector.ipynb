```python
import pandas as pd
import numpy as np
```

```python
# Vector Operations Project

# Define two vectors
v1 = np.array([3, 4])
v2 = np.array([1, 2])

print("Vector 1:", v1)
print("Vector 2:", v2)
print("\n" + "="*50)

# 1. Vector Addition
v_add = v1 + v2
print("\n1. Vector Addition:")
print(f"   {v1} + {v2} = {v_add}")

# 2. Vector Subtraction
v_sub = v1 - v2
print("\n2. Vector Subtraction:")
print(f"   {v1} - {v2} = {v_sub}")

# 3. Vector Length / Magnitude
magnitude_v1 = np.linalg.norm(v1)
magnitude_v2 = np.linalg.norm(v2)
print("\n3. Vector Magnitude:")
print(f"   ||{v1}|| = {magnitude_v1:.2f}")
print(f"   ||{v2}|| = {magnitude_v2:.2f}")

# 4. Linear Combination
# Linear combination: a*v1 + b*v2 where a and b are scalars
a, b = 2, 3
linear_comb = a * v1 + b * v2
print("\n4. Linear Combination:")
print(f"   {a}*{v1} + {b}*{v2} = {linear_comb}")

print("\n" + "="*50)
```

```python
# EXPLANATION: Linear Combination (a*v1 + b*v2)

print("LINEAR COMBINATION EXPLAINED")
print("="*60)

print("\nüìö What is a Linear Combination?")
print("-" * 60)
print("A linear combination means we multiply each vector by a scalar (number)")
print("and then add the results together.")
print("\nFormula: linear_comb = a*v1 + b*v2")
print(f"Where: a = {a}, b = {b}")
print(f"       v1 = {v1}, v2 = {v2}")

print("\nüî¢ Step-by-Step Calculation:")
print("-" * 60)
print(f"Step 1: Multiply v1 by a")
print(f"        {a} * {v1} = {a * v1}")

print(f"\nStep 2: Multiply v2 by b")
print(f"        {b} * {v2} = {b * v2}")

print(f"\nStep 3: Add the results")
print(f"        {a * v1} + {b * v2} = {linear_comb}")

print("\nüéØ Why multiply by 2 and 3?")
print("-" * 60)
print("The coefficients (a=2, b=3) are WEIGHTS that control:")
print("  ‚Ä¢ How much each vector contributes to the final result")
print("  ‚Ä¢ The relative importance of each vector")
print("  ‚Ä¢ The direction and magnitude of the combined vector")

print("\nüí° Practical Meaning:")
print("-" * 60)
print("‚Ä¢ a = 2 means: v1 contributes with 2x its original strength")
print("‚Ä¢ b = 3 means: v2 contributes with 3x its original strength")
print("‚Ä¢ Since b > a, vector v2 has MORE influence on the result")

print("\nüîç Visual Breakdown:")
print("-" * 60)
print(f"Original v1 contribution:     {v1}")
print(f"Scaled v1 (2x):               {a * v1}")
print(f"Original v2 contribution:     {v2}")
print(f"Scaled v2 (3x):               {b * v2}")
print(f"Combined result:              {linear_comb}")

print("\nüåü Real-World Applications:")
print("-" * 60)
print("‚Ä¢ Machine Learning: Combining features with different importance")
print("‚Ä¢ Physics: Forces acting in different directions")
print("‚Ä¢ Economics: Weighted portfolio of investments")
print("‚Ä¢ Fraud Detection: Weighting different transaction patterns")
print("\nIn our case, we might weight recent patterns (v2) more heavily")
print("than historical patterns (v1) to detect emerging fraud trends.")

print("\n" + "="*60)
```

```python
# SIMPLE FRAUD DETECTION SCENARIO: Credit Card Transaction Analysis
print("üè¶ CREDIT CARD FRAUD DETECTION SCENARIO")
print("="*70)

print("\nüìã SCENARIO:")
print("-" * 70)
print("A bank monitors credit card transactions using 2 key features:")
print("  ‚Ä¢ Feature 1: Transaction Amount (in $100s)")
print("  ‚Ä¢ Feature 2: Time Since Last Transaction (in hours)")
print()

# Define customer's normal behavior pattern
normal_pattern = np.array([5, 24])  # $500 every 24 hours (daily purchase)
print(f"üë§ Customer's Normal Pattern: {normal_pattern}")
print(f"   ‚Üí Typically spends ${normal_pattern[0]}00 every {normal_pattern[1]} hours")
print()

# Three different transactions to analyze
transaction_1 = np.array([6, 20])   # $600, 20 hours later - slightly unusual
transaction_2 = np.array([15, 2])   # $1500, 2 hours later - suspicious!
transaction_3 = np.array([4, 25])   # $400, 25 hours later - normal

print("üîç THREE TRANSACTIONS TO ANALYZE:")
print("-" * 70)
print(f"Transaction 1: {transaction_1} ‚Üí ${transaction_1[0]}00, {transaction_1[1]} hours later")
print(f"Transaction 2: {transaction_2} ‚Üí ${transaction_2[0]}00, {transaction_2[1]} hours later")
print(f"Transaction 3: {transaction_3} ‚Üí ${transaction_3[0]}00, {transaction_3[1]} hours later")
print()

# Analyze each transaction
print("\n‚öôÔ∏è  FRAUD DETECTION ANALYSIS:")
print("="*70)

for i, transaction in enumerate([transaction_1, transaction_2, transaction_3], 1):
    print(f"\n--- Transaction {i} ---")
    
    # Calculate deviation from normal
    deviation = transaction - normal_pattern
    print(f"Deviation from normal: {deviation}")
    print(f"  ‚Üí Amount diff: ${deviation[0]}00, Time diff: {deviation[1]} hours")
    
    # Calculate anomaly score (Euclidean distance)
    anomaly_score = np.linalg.norm(deviation)
    print(f"Anomaly Score: {anomaly_score:.2f}")
    
    # Decision threshold
    alert_threshold = 10.0
    if anomaly_score > alert_threshold:
        print(f"üö® FRAUD ALERT! Score {anomaly_score:.2f} > {alert_threshold}")
    else:
        print(f"‚úÖ APPROVED: Score {anomaly_score:.2f} ‚â§ {alert_threshold}")

print("\n" + "="*70)
print("\nüí° THE IMPORTANCE OF SCALARS IN FRAUD DETECTION:")
print("="*70)

print("\n1Ô∏è‚É£  WEIGHTING FEATURES BY IMPORTANCE")
print("-" * 70)
print("Different features have different importance in fraud detection.")
print("We use SCALARS to assign weights to each feature.\n")

# Define importance weights (scalars)
weight_amount = 3    # Amount is 3x more important
weight_time = 1      # Time is baseline importance

print(f"Weight for Amount: {weight_amount} (high risk indicator)")
print(f"Weight for Time: {weight_time} (moderate risk indicator)")
print("\nWhy? Large unusual amounts are stronger fraud signals than timing alone.\n")

# Apply weighted analysis to Transaction 2 (the suspicious one)
print("Analyzing Transaction 2 with WEIGHTED features:")
print(f"Original: {transaction_2}")

# Create weighted deviation
weighted_deviation = np.array([
    weight_amount * (transaction_2[0] - normal_pattern[0]),
    weight_time * (transaction_2[1] - normal_pattern[1])
])

weighted_score = np.linalg.norm(weighted_deviation)
unweighted_score = np.linalg.norm(transaction_2 - normal_pattern)

print(f"Unweighted anomaly score: {unweighted_score:.2f}")
print(f"Weighted anomaly score: {weighted_score:.2f}")
print(f"Impact: {((weighted_score/unweighted_score - 1) * 100):.1f}% increase in sensitivity")

print("\n2Ô∏è‚É£  SCALING FOR RISK LEVELS")
print("-" * 70)
print("Scalars help create RISK MULTIPLIERS for different scenarios:\n")

# Define risk multipliers
risk_scenarios = {
    'International': 2.5,
    'Late Night (1-5am)': 1.8,
    'New Merchant': 1.5,
    'Normal': 1.0
}

print("Risk Multipliers (Scalars):")
for scenario, multiplier in risk_scenarios.items():
    print(f"  {scenario}: {multiplier}x")

print(f"\nExample: Transaction 2 occurs internationally at 3am:")
scenario_multiplier = risk_scenarios['International'] * risk_scenarios['Late Night (1-5am)']
adjusted_score = weighted_score * scenario_multiplier
print(f"Base score: {weighted_score:.2f}")
print(f"Scenario multiplier: {scenario_multiplier:.2f}x")
print(f"Final adjusted score: {adjusted_score:.2f}")

if adjusted_score > 50:
    print("üö® CRITICAL ALERT: Block transaction immediately!")

print("\n3Ô∏è‚É£  THRESHOLD TUNING")
print("-" * 70)
print("Scalars control SENSITIVITY vs FALSE POSITIVES:\n")

# Different threshold levels
conservative_threshold = 5.0   # Catches more fraud, more false alarms
balanced_threshold = 10.0      # Balance between fraud and customer friction
permissive_threshold = 15.0    # Fewer false alarms, might miss some fraud

print(f"Conservative (strict): {conservative_threshold} ‚Üí Blocks more, safer")
print(f"Balanced: {balanced_threshold} ‚Üí Moderate approach")
print(f"Permissive (lenient): {permissive_threshold} ‚Üí Blocks less, riskier")

print("\n4Ô∏è‚É£  PRACTICAL EXAMPLE: Preventing a Real Fraud")
print("-" * 70)

# Simulate a fraudulent transaction pattern
fraud_transaction = np.array([50, 1])  # $5000, 1 hour after last purchase!
print(f"üè¥‚Äç‚ò†Ô∏è Fraudulent Transaction Attempt: {fraud_transaction}")
print(f"   ‚Üí ${fraud_transaction[0]}00, only {fraud_transaction[1]} hour later")
print()

# Apply full detection system with scalars
fraud_deviation = fraud_transaction - normal_pattern
print(f"Deviation: {fraud_deviation}")

# Apply feature weights
fraud_weighted = np.array([
    weight_amount * fraud_deviation[0],
    weight_time * fraud_deviation[1]
])
fraud_final_score = np.linalg.norm(fraud_weighted)

# Apply scenario multiplier (unusual time)
fraud_final_score *= risk_scenarios['Late Night (1-5am)']

print(f"Weighted fraud score: {fraud_final_score:.2f}")
print(f"Threshold: {balanced_threshold}")

if fraud_final_score > balanced_threshold:
    print("\n‚úã TRANSACTION BLOCKED!")
    print("   ‚Üí Customer receives SMS: 'Did you just try to spend $5000?'")
    print("   ‚Üí Fraud prevented, card temporarily frozen")
else:
    print("\n‚ö†Ô∏è  WARNING: May have been missed with lower sensitivity!")

print("\n" + "="*70)
print("üéØ KEY TAKEAWAYS ON SCALARS IN FRAUD DETECTION:")
print("="*70)
print("‚úì Scalars WEIGHT features by importance (amount > time)")
print("‚úì Scalars create RISK MULTIPLIERS for different scenarios")
print("‚úì Scalars set THRESHOLDS to balance security vs convenience")
print("‚úì Scalars AMPLIFY or REDUCE sensitivity to anomalies")
print("‚úì Without proper scalars, systems miss fraud or annoy customers!")
print("="*70)
```

```python
# FRAUD DETECTION USING VECTOR OPERATIONS
print("FRAUD DETECTION APPLICATION")
print("="*60)

# In fraud detection, each transaction can be represented as a vector
# where each dimension represents a different feature

# Example: Transaction vectors with features [amount, frequency]
# v1 represents a normal transaction pattern
# v2 represents another transaction to compare

print("\n1. TRANSACTION COMPARISON")
print("-" * 60)
# v1 = [3, 4] can represent: $300 transaction amount, 4 transactions/day
# v2 = [1, 2] can represent: $100 transaction amount, 2 transactions/day
print(f"Normal pattern (v1): Amount=${v1[0]}00, Frequency={v1[1]} trans/day")
print(f"New transaction (v2): Amount=${v2[0]}00, Frequency={v2[1]} trans/day")

print("\n2. DISTANCE-BASED ANOMALY DETECTION")
print("-" * 60)
# Calculate the difference between transaction patterns
# v_sub shows the deviation from normal behavior
print(f"Deviation vector: {v_sub}")
print(f"Explanation: Amount differs by ${v_sub[0]}00, Frequency by {v_sub[1]} trans/day")

# Calculate distance to determine if it's anomalous
distance = np.linalg.norm(v_sub)
print(f"\nDistance from normal pattern: {distance:.2f}")
# Smaller distance = more similar to normal behavior
# Larger distance = potential fraud
threshold = 3.0  # Example threshold
if distance > threshold:
    print(f"‚ö†Ô∏è  ALERT: Potential fraud detected (distance > {threshold})")
else:
    print(f"‚úì Transaction appears normal (distance ‚â§ {threshold})")

print("\n3. MAGNITUDE-BASED RISK SCORING")
print("-" * 60)
# Magnitude represents the overall "size" or "intensity" of a transaction
print(f"Normal transaction magnitude: {magnitude_v1:.2f}")
print(f"New transaction magnitude: {magnitude_v2:.2f}")
# Higher magnitude = higher value/frequency = higher risk in some contexts
risk_ratio = magnitude_v2 / magnitude_v1
print(f"Risk ratio: {risk_ratio:.2f}")
if risk_ratio > 1.5:
    print("‚ö†Ô∏è  High risk: Transaction significantly larger than normal")
elif risk_ratio < 0.5:
    print("‚ö†Ô∏è  Unusual: Transaction significantly smaller than normal")
else:
    print("‚úì Risk level: Normal")

print("\n4. WEIGHTED FRAUD SCORE (Linear Combination)")
print("-" * 60)
# linear_comb = 2*v1 + 3*v2 creates a weighted combination
# Weights (a=2, b=3) represent importance of different patterns
print(f"Weight for normal pattern (a): {a}")
print(f"Weight for comparison pattern (b): {b}")
print(f"Combined fraud score vector: {linear_comb}")
# Higher values indicate patterns that match known fraud profiles
fraud_score = np.linalg.norm(linear_comb)
print(f"Overall fraud score: {fraud_score:.2f}")
print("(Scores are compared against historical fraud patterns)")

print("\n5. PRACTICAL FRAUD DETECTION EXAMPLE")
print("-" * 60)
# Real-world example with actual transaction features
suspicious_transaction = np.array([10, 8])  # $1000, 8 trans/day (unusual)
print(f"Suspicious transaction: Amount=${suspicious_transaction[0]}00, "
      f"Frequency={suspicious_transaction[1]} trans/day")

# Calculate how different it is from normal
difference = suspicious_transaction - v1
anomaly_distance = np.linalg.norm(difference)
print(f"Anomaly distance from normal: {anomaly_distance:.2f}")

# Set decision threshold
fraud_threshold = 8.0
if anomaly_distance > fraud_threshold:
    print(f"üö® FRAUD ALERT: Distance {anomaly_distance:.2f} exceeds threshold {fraud_threshold}")
    print("   Recommended action: Block transaction and verify with customer")
else:
    print(f"‚úì Transaction approved: Distance {anomaly_distance:.2f} within threshold {fraud_threshold}")

print("\n" + "="*60)
print("KEY TAKEAWAYS:")
print("‚Ä¢ Vector subtraction ‚Üí measures deviation from normal")
print("‚Ä¢ Vector magnitude ‚Üí measures transaction intensity/size")
print("‚Ä¢ Distance calculation ‚Üí identifies anomalies")
print("‚Ä¢ Linear combination ‚Üí creates weighted fraud scores")
print("="*60)
```

```python
# DOT PRODUCT FUNDAMENTALS & PROPERTIES FOR AML/FRAUD DETECTION
print("="*80)
print("DOT PRODUCT IN AML & FINANCIAL CRIME DETECTION")
print("="*80)

# ============================================================================
# PART 1: WHAT IS DOT PRODUCT? - Simple Introduction
# ============================================================================
print("\nüìê PART 1: UNDERSTANDING DOT PRODUCT")
print("="*80)

# Simple example vectors
customer_transactions = np.array([3, 5])  # [number of wire transfers, number of cash deposits]
risk_weights = np.array([7, 4])            # [risk weight for wires, risk weight for cash]

print("Example: Customer Transaction Analysis")
print(f"Customer activity: {customer_transactions}")
print(f"  ‚Üí {customer_transactions[0]} wire transfers")
print(f"  ‚Üí {customer_transactions[1]} cash deposits")
print()
print(f"Risk weights: {risk_weights}")
print(f"  ‚Üí Wire transfers have risk weight: {risk_weights[0]}")
print(f"  ‚Üí Cash deposits have risk weight: {risk_weights[1]}")

# Calculate dot product manually
risk_score = np.dot(customer_transactions, risk_weights)
print(f"\nüî¢ DOT PRODUCT CALCULATION:")
print(f"Risk Score = (3 √ó 7) + (5 √ó 4)")
print(f"           = {3*7} + {5*4}")
print(f"           = {risk_score}")

print(f"\nüí° MEANING: This customer has a total risk score of {risk_score}")
print("   Higher score ‚Üí Higher money laundering risk")

# ============================================================================
# PART 2: DISTRIBUTIVE PROPERTY - Real AML Scenario
# ============================================================================
print("\n" + "="*80)
print("üìö PART 2: DISTRIBUTIVE PROPERTY IN AML")
print("="*80)
print("\nüîç PROPERTY: weights ¬∑ (account1 + account2) = weights ¬∑ account1 + weights ¬∑ account2")
print("   Translation: 'Analyze accounts separately OR combined - same result!'")
print("-"*80)

print("\nüè¶ SCENARIO: Customer with Multiple Accounts")
print("-"*80)

# AML weights for different transaction types
aml_weights = np.array([8, 5, 6])  # [international_wires, crypto_purchases, large_cash]

# Customer has 2 accounts with different activity
account_1 = np.array([2, 1, 0])  # Business account: 2 int'l wires, 1 crypto, 0 cash
account_2 = np.array([1, 3, 2])  # Personal account: 1 int'l wire, 3 crypto, 2 large cash

print("AML Risk Weights: [International Wires: 8, Crypto: 5, Large Cash: 6]")
print(f"\nAccount 1 (Business): {account_1}")
print(f"  ‚Üí {account_1[0]} international wires")
print(f"  ‚Üí {account_1[1]} crypto purchases")
print(f"  ‚Üí {account_1[2]} large cash transactions")

print(f"\nAccount 2 (Personal): {account_2}")
print(f"  ‚Üí {account_2[0]} international wire")
print(f"  ‚Üí {account_2[1]} crypto purchases")
print(f"  ‚Üí {account_2[2]} large cash transactions")

# METHOD 1: Combine accounts first, then calculate risk
print("\n" + "-"*80)
print("METHOD 1Ô∏è‚É£: Combine All Accounts First (Holistic View)")
print("-"*80)
combined_accounts = account_1 + account_2
print(f"Step 1: Combine accounts: {account_1} + {account_2} = {combined_accounts}")
print(f"        Total activity across both accounts: {combined_accounts}")

total_risk_method1 = np.dot(aml_weights, combined_accounts)
print(f"\nStep 2: Calculate total risk score")
print(f"        {aml_weights} ¬∑ {combined_accounts}")
print(f"        = (8√ó{combined_accounts[0]}) + (5√ó{combined_accounts[1]}) + (6√ó{combined_accounts[2]})")
print(f"        = {8*combined_accounts[0]} + {5*combined_accounts[1]} + {6*combined_accounts[2]}")
print(f"        = {total_risk_method1}")

# METHOD 2: Calculate risk per account, then add
print("\n" + "-"*80)
print("METHOD 2Ô∏è‚É£: Analyze Each Account Separately")
print("-"*80)
risk_account1 = np.dot(aml_weights, account_1)
risk_account2 = np.dot(aml_weights, account_2)

print(f"Step 1: Calculate Account 1 risk")
print(f"        {aml_weights} ¬∑ {account_1}")
print(f"        = (8√ó2) + (5√ó1) + (6√ó0) = {risk_account1}")

print(f"\nStep 2: Calculate Account 2 risk")
print(f"        {aml_weights} ¬∑ {account_2}")
print(f"        = (8√ó1) + (5√ó3) + (6√ó2) = {risk_account2}")

total_risk_method2 = risk_account1 + risk_account2
print(f"\nStep 3: Add individual risks")
print(f"        {risk_account1} + {risk_account2} = {total_risk_method2}")

print(f"\n‚úÖ VERIFICATION: {total_risk_method1} = {total_risk_method2} ‚úì")
print("   DISTRIBUTIVE PROPERTY CONFIRMED!")

print(f"\nüö® AML DECISION: Total risk score = {total_risk_method1}")
if total_risk_method1 > 50:
    print("   ‚Üí HIGH RISK: File Suspicious Activity Report (SAR)")
elif total_risk_method1 > 30:
    print("   ‚Üí MEDIUM RISK: Enhanced monitoring required")
else:
    print("   ‚Üí LOW RISK: Standard monitoring")

# ============================================================================
# PART 3: ASSOCIATIVE PROPERTY - Adjusting Detection Sensitivity
# ============================================================================
print("\n" + "="*80)
print("üìö PART 3: ASSOCIATIVE PROPERTY IN FRAUD DETECTION")
print("="*80)
print("\nüîç PROPERTY: (sensitivity √ó weights) ¬∑ activity = sensitivity √ó (weights ¬∑ activity)")
print("   Translation: 'Can adjust sensitivity by scaling weights OR final score!'")
print("-"*80)

print("\nüí≥ SCENARIO: Credit Card Fraud Detection - Adjusting Sensitivity")
print("-"*80)

# Transaction features
suspicious_activity = np.array([1, 1, 0, 1])  # [foreign_country, unusual_time, new_merchant, high_amount]
base_weights = np.array([6, 4, 3, 7])          # Base risk weights

sensitivity_multiplier = 1.5  # Increase sensitivity by 50% during holiday season

print("Transaction Features (1=yes, 0=no):")
print(f"  Suspicious activity: {suspicious_activity}")
print(f"  ‚Üí Foreign country: {suspicious_activity[0]}")
print(f"  ‚Üí Unusual time (3am): {suspicious_activity[1]}")
print(f"  ‚Üí New merchant: {suspicious_activity[2]}")
print(f"  ‚Üí High amount: {suspicious_activity[3]}")

print(f"\nBase weights: {base_weights}")
print(f"Sensitivity multiplier: {sensitivity_multiplier}x (Holiday fraud increase)")

# METHOD 1: Scale the weights first
print("\n" + "-"*80)
print("METHOD 1Ô∏è‚É£: Scale Weights First (Adjust Detection Rules)")
print("-"*80)
scaled_weights = sensitivity_multiplier * base_weights
print(f"Step 1: Scale weights")
print(f"        {sensitivity_multiplier} √ó {base_weights} = {scaled_weights}")

fraud_score_method1 = np.dot(scaled_weights, suspicious_activity)
print(f"\nStep 2: Calculate fraud score")
print(f"        {scaled_weights} ¬∑ {suspicious_activity}")
print(f"        = {fraud_score_method1}")

# METHOD 2: Calculate base score first, then scale
print("\n" + "-"*80)
print("METHOD 2Ô∏è‚É£: Calculate Base Score First, Then Adjust")
print("-"*80)
base_fraud_score = np.dot(base_weights, suspicious_activity)
print(f"Step 1: Calculate base fraud score")
print(f"        {base_weights} ¬∑ {suspicious_activity}")
print(f"        = {base_fraud_score}")

fraud_score_method2 = sensitivity_multiplier * base_fraud_score
print(f"\nStep 2: Apply sensitivity multiplier")
print(f"        {sensitivity_multiplier} √ó {base_fraud_score} = {fraud_score_method2}")

# METHOD 3: Scale the activity vector
print("\n" + "-"*80)
print("METHOD 3Ô∏è‚É£: Scale Activity Features (Amplify Signals)")
print("-"*80)
scaled_activity = sensitivity_multiplier * suspicious_activity
print(f"Step 1: Scale activity signals")
print(f"        {sensitivity_multiplier} √ó {suspicious_activity} = {scaled_activity}")

fraud_score_method3 = np.dot(base_weights, scaled_activity)
print(f"\nStep 2: Calculate fraud score with amplified signals")
print(f"        {base_weights} ¬∑ {scaled_activity}")
print(f"        = {fraud_score_method3}")

print(f"\n‚úÖ VERIFICATION: {fraud_score_method1} = {fraud_score_method2} = {fraud_score_method3} ‚úì")
print("   ASSOCIATIVE PROPERTY CONFIRMED!")

fraud_threshold = 20
print(f"\nüö® FRAUD DECISION: Score = {fraud_score_method1}, Threshold = {fraud_threshold}")
if fraud_score_method1 > fraud_threshold:
    print("   ‚Üí BLOCK TRANSACTION & Alert customer")
else:
    print("   ‚Üí APPROVE transaction")

# ============================================================================
# PART 4: COMBINED REAL-WORLD AML EXAMPLE
# ============================================================================
print("\n" + "="*80)
print("üè¶ PART 4: COMPLETE AML SCENARIO - Money Laundering Detection")
print("="*80)

print("\nüìã BACKGROUND:")
print("-"*80)
print("A bank monitors 3 customers for potential money laundering.")
print("Each customer has activity across multiple transaction types.")
print("The bank needs to calculate overall risk efficiently.")

# Define transaction types and their AML risk weights
transaction_types = ['Cash Deposits >$10k', 'Wire to High-Risk Country', 
                     'Shell Company Payments', 'Crypto Conversions']
aml_risk_weights = np.array([6, 9, 10, 7])

# Customer activities (number of each transaction type)
customer_A = np.array([2, 1, 0, 1])  # Moderate risk
customer_B = np.array([0, 0, 3, 2])  # High risk (shell companies)
customer_C = np.array([1, 0, 0, 0])  # Low risk

print(f"\nAML Risk Weights: {aml_risk_weights}")
for i, tx_type in enumerate(transaction_types):
    print(f"  {tx_type}: {aml_risk_weights[i]}")

print(f"\nCustomer A activity: {customer_A}")
print(f"Customer B activity: {customer_B}")
print(f"Customer C activity: {customer_C}")

# Using DISTRIBUTIVE property for portfolio analysis
print("\n" + "-"*80)
print("üîπ DISTRIBUTIVE PROPERTY: Total Portfolio Risk")
print("-"*80)

# Method 1: Combined
total_activity = customer_A + customer_B + customer_C
portfolio_risk = np.dot(aml_risk_weights, total_activity)
print(f"Combined activity: {total_activity}")
print(f"Total portfolio risk: {portfolio_risk}")

# Method 2: Individual then sum
risk_A = np.dot(aml_risk_weights, customer_A)
risk_B = np.dot(aml_risk_weights, customer_B)
risk_C = np.dot(aml_risk_weights, customer_C)
portfolio_risk_sum = risk_A + risk_B + risk_C

print(f"\nIndividual risks:")
print(f"  Customer A: {risk_A}")
print(f"  Customer B: {risk_B} ‚ö†Ô∏è HIGHEST")
print(f"  Customer C: {risk_C}")
print(f"  Sum: {portfolio_risk_sum}")

print(f"\n‚úÖ Both methods give {portfolio_risk} - Distributive property!")

# Using ASSOCIATIVE property for regulatory changes
print("\n" + "-"*80)
print("üîπ ASSOCIATIVE PROPERTY: Regulatory Alert Adjustment")
print("-"*80)

print("\nüì¢ ALERT: New regulation increases shell company risk by 1.5x")
regulatory_multiplier = 1.5

# Analyze Customer B under new regulations
print(f"\nCustomer B (highest risk) under new rules:")
print(f"Original risk score: {risk_B}")

# Can adjust either the specific weight or the final score
adjusted_risk = regulatory_multiplier * risk_B
print(f"Adjusted risk: {regulatory_multiplier} √ó {risk_B} = {adjusted_risk}")

critical_threshold = 50
if adjusted_risk > critical_threshold:
    print(f"\nüö® CRITICAL: Score {adjusted_risk} > {critical_threshold}")
    print("   ACTIONS REQUIRED:")
    print("   ‚úì File Suspicious Activity Report (SAR)")
    print("   ‚úì Freeze account pending investigation")
    print("   ‚úì Notify compliance officer")
    print("   ‚úì Request enhanced due diligence")

# ============================================================================
# SUMMARY & KEY TAKEAWAYS
# ============================================================================
print("\n" + "="*80)
print("üéØ KEY TAKEAWAYS FOR AML/FRAUD DETECTION")
print("="*80)

print("\n‚úÖ DOT PRODUCT:")
print("   ‚Ä¢ Combines multiple risk factors into a single score")
print("   ‚Ä¢ Each factor weighted by its importance")
print("   ‚Ä¢ Formula: score = w‚ÇÅ√óf‚ÇÅ + w‚ÇÇ√óf‚ÇÇ + ... + w‚Çô√óf‚Çô")

print("\n‚úÖ DISTRIBUTIVE PROPERTY: weights¬∑(A + B) = weights¬∑A + weights¬∑B")
print("   ‚Ä¢ Analyze customers separately OR combined - same result")
print("   ‚Ä¢ Useful for portfolio-level risk assessment")
print("   ‚Ä¢ Enables modular risk calculation systems")

print("\n‚úÖ ASSOCIATIVE PROPERTY: (c√óweights)¬∑activity = c√ó(weights¬∑activity)")
print("   ‚Ä¢ Adjust sensitivity by scaling weights OR scores")
print("   ‚Ä¢ Adapt to regulatory changes efficiently")
print("   ‚Ä¢ Control false positive vs false negative rates")

print("\nüíº PRACTICAL APPLICATIONS:")
print("   üè¶ AML: Aggregate risk across multiple accounts/customers")
print("   üí≥ Fraud: Adjust detection sensitivity for special periods")
print("   üìä Compliance: Respond quickly to new regulations")
print("   üéØ Investigation: Prioritize high-risk cases efficiently")

print("\n" + "="*80)
```

```python
# DOT PRODUCT: DISTRIBUTIVE & ASSOCIATIVE PROPERTIES
print("="*70)
print("DOT PRODUCT PROPERTIES WITH SIMPLE EXAMPLES")
print("="*70)

# ============================================================================
# PART 1: WHAT IS DOT PRODUCT?
# ============================================================================
print("\nüìê PART 1: UNDERSTANDING DOT PRODUCT")
print("="*70)

# Simple vectors
u = np.array([2, 3])
v = np.array([4, 1])
w = np.array([1, 2])

print("Vectors:")
print(f"u = {u}")
print(f"v = {v}")
print(f"w = {w}")

# Calculate dot product
dot_uv = np.dot(u, v)
print(f"\nDot Product u¬∑v = (2)(4) + (3)(1) = {2*4} + {3*1} = {dot_uv}")
print("\nüí° Dot product = multiply corresponding elements and add them up")

# ============================================================================
# PART 2: DISTRIBUTIVE PROPERTY OVER ADDITION
# ============================================================================
print("\n" + "="*70)
print("üìö PART 2: DISTRIBUTIVE PROPERTY OVER ADDITION")
print("="*70)
print("\nüîç Property: u¬∑(v + w) = u¬∑v + u¬∑w")
print("Translation: 'Dot product distributes over vector addition'")
print("-"*70)

print("\nMethod 1Ô∏è‚É£: Calculate u¬∑(v + w) directly")
print("-"*70)
v_plus_w = v + w
print(f"Step 1: v + w = {v} + {w} = {v_plus_w}")
left_side = np.dot(u, v_plus_w)
print(f"Step 2: u¬∑(v + w) = {u}¬∑{v_plus_w}")
print(f"        = (2)(5) + (3)(3) = {2*5} + {3*3} = {left_side}")

print("\nMethod 2Ô∏è‚É£: Calculate u¬∑v + u¬∑w separately")
print("-"*70)
dot_u_v = np.dot(u, v)
dot_u_w = np.dot(u, w)
print(f"Step 1: u¬∑v = {u}¬∑{v} = (2)(4) + (3)(1) = {dot_u_v}")
print(f"Step 2: u¬∑w = {u}¬∑{w} = (2)(1) + (3)(2) = {dot_u_w}")
right_side = dot_u_v + dot_u_w
print(f"Step 3: u¬∑v + u¬∑w = {dot_u_v} + {dot_u_w} = {right_side}")

print(f"\n‚úÖ VERIFICATION: {left_side} = {right_side} ‚úì")
print("The distributive property holds!")

# Real-world example
print("\n" + "="*70)
print("üõí REAL-WORLD EXAMPLE: Shopping Cart")
print("="*70)

prices = np.array([10, 5])  # [price_apples, price_oranges]
cart1 = np.array([3, 2])    # [3 apples, 2 oranges]
cart2 = np.array([1, 4])    # [1 apple, 4 oranges]

print(f"Prices: Apples=${prices[0]}, Oranges=${prices[1]}")
print(f"Cart 1: {cart1[0]} apples, {cart1[1]} oranges")
print(f"Cart 2: {cart2[0]} apple, {cart2[1]} oranges")

# Two ways to calculate total cost
total_cart = cart1 + cart2
cost_combined = np.dot(prices, total_cart)
print(f"\nMethod 1: Combine carts first, then calculate cost")
print(f"  Total items: {total_cart}")
print(f"  Total cost: ${cost_combined}")

cost1 = np.dot(prices, cart1)
cost2 = np.dot(prices, cart2)
cost_separate = cost1 + cost2
print(f"\nMethod 2: Calculate costs separately, then add")
print(f"  Cart 1 cost: ${cost1}")
print(f"  Cart 2 cost: ${cost2}")
print(f"  Total cost: ${cost_separate}")

print(f"\n‚úÖ Both methods give ${cost_combined} - Distributive property in action!")

# ============================================================================
# PART 3: ASSOCIATIVE PROPERTY OVER SCALAR MULTIPLICATION
# ============================================================================
print("\n" + "="*70)
print("üìö PART 3: ASSOCIATIVE PROPERTY OVER SCALAR MULTIPLICATION")
print("="*70)
print("\nüîç Property: (c¬∑u)¬∑v = c¬∑(u¬∑v) = u¬∑(c¬∑v)")
print("Translation: 'Scalar can be moved around in dot product'")
print("-"*70)

c = 3  # scalar
print(f"\nScalar: c = {c}")
print(f"Vectors: u = {u}, v = {v}")

print("\nMethod 1Ô∏è‚É£: (c¬∑u)¬∑v - Scale u first, then dot product")
print("-"*70)
c_times_u = c * u
print(f"Step 1: c¬∑u = {c}¬∑{u} = {c_times_u}")
result1 = np.dot(c_times_u, v)
print(f"Step 2: (c¬∑u)¬∑v = {c_times_u}¬∑{v}")
print(f"        = (6)(4) + (9)(1) = {6*4} + {9*1} = {result1}")

print("\nMethod 2Ô∏è‚É£: c¬∑(u¬∑v) - Dot product first, then scale")
print("-"*70)
u_dot_v = np.dot(u, v)
print(f"Step 1: u¬∑v = {u}¬∑{v} = {u_dot_v}")
result2 = c * u_dot_v
print(f"Step 2: c¬∑(u¬∑v) = {c}¬∑{u_dot_v} = {result2}")

print("\nMethod 3Ô∏è‚É£: u¬∑(c¬∑v) - Scale v first, then dot product")
print("-"*70)
c_times_v = c * v
print(f"Step 1: c¬∑v = {c}¬∑{v} = {c_times_v}")
result3 = np.dot(u, c_times_v)
print(f"Step 2: u¬∑(c¬∑v) = {u}¬∑{c_times_v}")
print(f"        = (2)(12) + (3)(3) = {2*12} + {3*3} = {result3}")

print(f"\n‚úÖ VERIFICATION: {result1} = {result2} = {result3} ‚úì")
print("The associative property holds!")

# Real-world example
print("\n" + "="*70)
print("üí∞ REAL-WORLD EXAMPLE: Investment Returns")
print("="*70)

portfolio = np.array([100, 200])  # [stocks, bonds] in units
returns = np.array([0.10, 0.05])  # [10% stock return, 5% bond return]
years = 3

print(f"Portfolio: {portfolio[0]} stocks, {portfolio[1]} bonds")
print(f"Annual returns: {returns[0]*100}% stocks, {returns[1]*100}% bonds")
print(f"Time period: {years} years")

print("\nMethod 1: Calculate annual return, then multiply by years")
print("-"*70)
annual_return = np.dot(portfolio, returns)
total_return1 = years * annual_return
print(f"  Annual return: ${annual_return:.2f}")
print(f"  Total ({years} years): ${total_return1:.2f}")

print("\nMethod 2: Scale portfolio by years, then calculate return")
print("-"*70)
scaled_portfolio = years * portfolio
total_return2 = np.dot(scaled_portfolio, returns)
print(f"  Scaled portfolio: {scaled_portfolio}")
print(f"  Total return: ${total_return2:.2f}")

print("\nMethod 3: Scale returns by years, then calculate")
print("-"*70)
scaled_returns = years * returns
total_return3 = np.dot(portfolio, scaled_returns)
print(f"  Scaled returns: {scaled_returns}")
print(f"  Total return: ${total_return3:.2f}")

print(f"\n‚úÖ All methods give ${total_return1:.2f} - Associative property!")

# ============================================================================
# PART 4: FRAUD DETECTION APPLICATION
# ============================================================================
print("\n" + "="*70)
print("üîç PART 4: FRAUD DETECTION WITH DOT PRODUCT PROPERTIES")
print("="*70)

# Feature vectors
high_risk_features = np.array([1, 1, 1, 0])  # [international, late_night, new_merchant, large_amount]
med_risk_features = np.array([0, 1, 1, 0])   # [domestic, late_night, new_merchant, normal_amount]

# Risk weights (how important each feature is)
risk_weights = np.array([5, 3, 2, 4])  # importance scores

print("\nRisk Features (1=yes, 0=no):")
print(f"  High-risk transaction: {high_risk_features}")
print(f"  Medium-risk transaction: {med_risk_features}")
print(f"\nRisk Weights: {risk_weights}")
print(f"  [International: 5, Late Night: 3, New Merchant: 2, Large Amount: 4]")

# Using distributive property
print("\nüîπ DISTRIBUTIVE PROPERTY APPLICATION")
print("-"*70)
print("Question: What's the risk if we combine both transactions?")

combined_features = high_risk_features + med_risk_features
combined_risk = np.dot(risk_weights, combined_features)
print(f"\nMethod 1: Combine features first")
print(f"  Combined: {combined_features}")
print(f"  Risk score: {combined_risk}")

high_risk_score = np.dot(risk_weights, high_risk_features)
med_risk_score = np.dot(risk_weights, med_risk_features)
total_risk = high_risk_score + med_risk_score
print(f"\nMethod 2: Calculate risks separately")
print(f"  High-risk score: {high_risk_score}")
print(f"  Medium-risk score: {med_risk_score}")
print(f"  Total risk: {total_risk}")

print(f"\n‚úÖ Both give risk score of {combined_risk} - Distributive property!")

# Using associative property
print("\nüîπ ASSOCIATIVE PROPERTY APPLICATION")
print("-"*70)
print("Question: What if we increase sensitivity by 2x?")

sensitivity = 2

method1 = np.dot(sensitivity * risk_weights, high_risk_features)
method2 = sensitivity * np.dot(risk_weights, high_risk_features)
method3 = np.dot(risk_weights, sensitivity * high_risk_features)

print(f"\nMethod 1: (2 √ó weights) ¬∑ features = {method1}")
print(f"Method 2: 2 √ó (weights ¬∑ features) = {method2}")
print(f"Method 3: weights ¬∑ (2 √ó features) = {method3}")

print(f"\n‚úÖ All give {method1} - Associative property!")
print("   We can adjust sensitivity by scaling weights OR features!")

# ============================================================================
# SUMMARY
# ============================================================================
print("\n" + "="*70)
print("üéØ KEY TAKEAWAYS")
print("="*70)
print("\n‚úì DISTRIBUTIVE: u¬∑(v + w) = u¬∑v + u¬∑w")
print("  ‚Üí Useful when combining multiple vectors before calculation")
print("  ‚Üí Example: Total cost of combined shopping carts")
print()
print("‚úì ASSOCIATIVE: (c¬∑u)¬∑v = c¬∑(u¬∑v) = u¬∑(c¬∑v)")
print("  ‚Üí Scalar can move freely in the calculation")
print("  ‚Üí Example: Scaling sensitivity in fraud detection")
print()
print("‚úì FRAUD DETECTION:")
print("  ‚Üí Distributive: Analyze combined transaction patterns efficiently")
print("  ‚Üí Associative: Adjust risk sensitivity without recalculating features")
print("="*70)
```
